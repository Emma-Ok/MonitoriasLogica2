# Acercamiento a Matplotlib.

*¿Qué es Matplotlib?* Matplotlib es una librería de Python utilizada para crear gráficos estáticos, animados e interactivos. Es una de las herramientas más poderosas y flexibles en el ecosistema de visualización de datos de Python. Con Matplotlib, puedes crear todo tipo de gráficos, desde simples gráficos de líneas hasta complejos gráficos en 3D.

*¿Por qué es útil?* Matplotlib es especialmente útil en el análisis de datos, ya que permite visualizar grandes volúmenes de información de manera comprensible. Ya sea en ciencia de datos, machine learning o análisis financiero, la capacidad de crear gráficos claros y personalizables ayuda a comunicar resultados de forma efectiva.



# Estructura básica de Matplotlib

*Componentes Principales*
  * pyplot: Es el módulo más utilizado en Matplotlib. Proporciona una interfaz simple y fácil de usar para crear gráficos rápidamente
  * figure: Es el contenedor principal de un gráfico. Representa la ventana o el espacio donde se dibuja el gráfico.
  * axes: Son los ejes del gráfico y contienen los datos visualizados. Cada figura puede tener múltiples ejes.

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos
x = np.linspace(0, 4, 100) # np.linspace nos da valores desde un rango especificado.
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.random.rand(100)  # Datos aleatorios para el scatter plot

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Dibujar líneas y scatter
ax.plot(x, y1, label='Blue signal', color='blue')
ax.plot(x, y2, label='Orange signal', color='orange')
ax.scatter(x, y3, label='Markers', facecolors='none', edgecolors='purple', marker='s') # filled_markers = ('.', 'o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X')

# Título del gráfico
ax.set_title('Ejemplo de figura')

# Etiquetas de los ejes
ax.set_xlabel('x Eje')
ax.set_ylabel('y Eje')

# Leyenda
ax.legend()

# Configuración de la grilla
ax.grid(True)

# Configuración de las spines (bordes del gráfico)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Configuración de los ticks mayores y menores
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.25))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: '{:.0f}'.format(val)))

ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.25))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: '{:.0f}'.format(val)))

# Mostrar el gráfico
plt.show()
```

```python
from typing import List
class AlgoritmosOrdenamiento:
    @staticmethod
    def bubble_sort(arr: List[int]) -> List[int]:
        n = len(arr)
        arr_copy = arr.copy()
        for i in range(n):
            for j in range(0, n - i - 1):
                if arr_copy[j] > arr_copy[j + 1]:
                    arr_copy[j], arr_copy[j + 1] = arr_copy[j + 1], arr_copy[j]
        return arr_copy

    @staticmethod
    def selection_sort(arr: List[int]) -> List[int]:
        arr_copy = arr.copy()
        for i in range(len(arr_copy)):
            min_idx = i
            for j in range(i + 1, len(arr_copy)):
                if arr_copy[j] < arr_copy[min_idx]:
                    min_idx = j
            arr_copy[i], arr_copy[min_idx] = arr_copy[min_idx], arr_copy[i]
        return arr_copy

    @staticmethod
    def insertion_sort(arr: List[int]) -> List[int]:
        arr_copy = arr.copy()
        for i in range(1, len(arr_copy)):
            key = arr_copy[i]
            j = i - 1
            while j >= 0 and arr_copy[j] > key:
                arr_copy[j + 1] = arr_copy[j]
                j -= 1
            arr_copy[j + 1] = key
        return arr_copy

    @staticmethod
    def merge_sort(arr: List[int]) -> List[int]:
        if len(arr) <= 1:
            return arr

        mid = len(arr) // 2
        left = AlgoritmosOrdenamiento.merge_sort(arr[:mid])
        right = AlgoritmosOrdenamiento.merge_sort(arr[mid:])

        return AlgoritmosOrdenamiento._merge(left, right)

    @staticmethod
    def _merge(left: List[int], right: List[int]) -> List[int]:
        result = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1

        result.extend(left[i:])
        result.extend(right[j:])
        return result

    @staticmethod
    def quick_sort(arr: List[int]) -> List[int]:
        if len(arr) <= 1:
            return arr
        else:
            pivot = arr[0]
            less = [x for x in arr[1:] if x <= pivot]
            greater = [x for x in arr[1:] if x > pivot]
            return AlgoritmosOrdenamiento.quick_sort(less) + [pivot] + AlgoritmosOrdenamiento.quick_sort(greater)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def medir_tiempo_ordenamiento(algoritmo, arr):
    inicio = time.time()
    algoritmo(arr)
    fin = time.time()
    return fin - inicio

# Configuración de la visualización
plt.style.use('seaborn')
fig = plt.figure(figsize=(20, 12))

# Lista de algoritmos a evaluar
algoritmos = {
    'Bubble Sort': AlgoritmosOrdenamiento.bubble_sort,
    'Selection Sort': AlgoritmosOrdenamiento.selection_sort,
    'Insertion Sort': AlgoritmosOrdenamiento.insertion_sort,
    'Merge Sort': AlgoritmosOrdenamiento.merge_sort,
    'Quick Sort': AlgoritmosOrdenamiento.quick_sort
}

# Definir tres rangos diferentes de tamaños para mejor visualización
small_sizes = np.linspace(100, 1000, 10, dtype=int)
medium_sizes = np.linspace(1000, 2000, 10, dtype=int)
large_sizes = np.linspace(2000, 5000, 5, dtype=int)

# Tamaños de arrays a probar
all_sizes = np.concatenate([small_sizes, medium_sizes, large_sizes])
tiempos = {nombre: [] for nombre in algoritmos}

# Medir tiempos para cada algoritmo y tamaño
for size in all_sizes:
    arr = np.random.randint(0, 10, size=size)
    for nombre, algoritmo in algoritmos.items():
        tiempo = medir_tiempo_ordenamiento(algoritmo, arr.tolist())
        tiempos[nombre].append(tiempo)

fig1 = plt.figure(figsize=(20, 10))
fig1.suptitle('Análisis Individual de Algoritmos de Ordenamiento', fontsize=16)
# Crear subplots individuales
for idx, (nombre, algoritmo) in enumerate(algoritmos.items(), 1):
    plt.subplot(2, 3, idx)
    plt.plot(all_sizes, tiempos[nombre], '-o')
    plt.title(f'{nombre}', fontsize=12)
    plt.xlabel('Tamaño del array', fontsize=10)
    plt.ylabel('Tiempo (segundos)', fontsize=10)
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Crear figura para gráfico comparativo
fig2 = plt.figure(figsize=(15, 8))
for nombre, algo_info in algoritmos.items():
    plt.plot(all_sizes, tiempos[nombre], '-o', label=nombre)
plt.title('Comparación de Algoritmos de Ordenamiento', fontsize=14)
plt.xlabel('Tamaño del array', fontsize=12)
plt.ylabel('Tiempo (segundos)', fontsize=12)
plt.legend(fontsize=10, loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Análisis de resultados
print("Análisis de rendimiento:")
for size in [1000, 2000, 5000]:
    print(f"\nPara arrays de tamaño {size}:")
    idx = np.where(all_sizes == size)[0][0]
    sorted_algorithms = sorted(algoritmos.keys(), key=lambda x: tiempos[x][idx])
    for i, algo in enumerate(sorted_algorithms, 1):
        print(f"{i}. {algo}: {tiempos[algo][idx]:.6f} segundos")
```
